# Lab 4: Simulations for Central Limit Theorem (CLT)

```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Example 1: IID Samples from Uniform Distribution

```python
x = np.linspace(-0.5, 1.5, 501)


plt.fill_between(x, scipy.stats.uniform().pdf(x), alpha = 0.3)
plt.legend(["Uniform(0, 1)"])
plt.grid(True)
```

```python
N = 1

while N < 100: 
    sample_means = [scipy.stats.uniform().rvs(N).mean() for _ in range(1000)]
    plt.hist(sample_means, bins = 41, alpha = 0.3)
    print("Sample size = {0} and {1} H_0.".format(N, "reject" if scipy.stats.shapiro(sample_means)[1] < 0.05 else "not reject"))
    N *= 2
    
    
plt.grid(True)
```

## Example 2: IID Samples from $\chi^2$ Distribution

```python
x = np.linspace(0, 10, 101)


plt.fill_between(x, scipy.stats.chi2(df = 3).pdf(x), alpha = 0.3)
plt.grid(True)
plt.legend(["$\chi^2$(3)"])
```

```python
N = 1

while N < 1000: 
    sample_means = [scipy.stats.chi2(df = 3).rvs(N).mean() for _ in range(1000)]
    plt.hist(sample_means, bins = 41, alpha = 0.3)
    print("Sample size = {0} and {1} H_0.".format(N, "reject" if scipy.stats.shapiro(sample_means)[1] < 0.05 else "not reject"))
    N *= 2
    
    
plt.grid(True)
```

## Example 3: IID Samples from Poisson Distribution

```python
x = np.linspace(0, 10, 11)
y = scipy.stats.poisson(mu = 3).pmf(x)


plt.bar(x, y, alpha = 0.6, width = 0.3)
plt.legend(["Poisson(3)"])
plt.grid(True)
```

```python
N = 1

while N < 1000: 
    sample_means = [scipy.stats.poisson(mu = 3).rvs(N).mean() for _ in range(1000)]
    plt.hist(sample_means, bins = 41, alpha = 0.3)
    print("Sample size = {0} and {1} H_0.".format(N, "reject" if scipy.stats.shapiro(sample_means)[1] < 0.05 else "not reject"))
    N *= 2
    
plt.grid(True)
```

## Example 4: IID Samples from Cauchy Distribution
- The Cauchy distribution, denoted by Cauchy($x_0$, $\gamma$), has the pdf $$f(x; x_0, \gamma) = \dfrac{1}{\pi \gamma} \dfrac{1}{\left(1 + (x - x_0) ^ 2 / \gamma ^ 2\right)}, $$ where $x_0$ is the location parameter and $\gamma$ is the scale parameter.
- Note that the variance of Cauchy($x_0$, $\gamma$) is **infinite**.
- References:
    - https://en.wikipedia.org/wiki/Cauchy_distribution



```python
x = np.linspace(-3, 3, 501)
y = scipy.stats.cauchy().pdf(x)
yn = scipy.stats.norm().pdf(x)


plt.fill_between(x, y, alpha = 0.3)
plt.plot(x, yn, alpha = 0.3, color = "orange")
plt.legend(["Normal(0, 1)", "Cauchy"])
plt.grid(True)
```

```python
N = 1

while N < 10000: 
    sample_means = [scipy.stats.cauchy().rvs(N).mean() for _ in range(1000)]
    plt.hist(sample_means, bins = 41, alpha = 0.3)
    print("Sample size = {0} and {1} H_0.".format(N, "reject" if scipy.stats.shapiro(sample_means)[1] < 0.05 else "not reject"))
    N *= 2
    
    
plt.grid(True)
```
