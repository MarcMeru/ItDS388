```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def rng(size = 1):

    x = scipy.stats.uniform().rvs(int(size))
    y = np.zeros((int(size), 1))
    for i in range(int(size)):
        if 0 < x[i] < 0.1:
            y[i] = -2
        elif 0.1 <= x[i] < 0.5:
            y[i] = 0
        elif 0.5 <= x[i] < 0.8:
            y[i] = 1
        elif 0.8 <= x[i] <= 1:
            y[i] = 4

    return y


sample_sizes = np.round(np.linspace(10, 3000, 1000))
samples = [rng(s) for s in sample_sizes]
means = list(map(np.mean, samples))


plt.figure(figsize = (10, 6))
plt.plot(sample_sizes, means, "o--", alpha = 0.1, mfc = 'g', mec = 'g')
plt.hlines(0.9, min(sample_sizes), max(sample_sizes), "r")
plt.grid(True)
plt.xlabel("Sample Size")
plt.legend(["Samples Mean", "Expectation Value"])
```
