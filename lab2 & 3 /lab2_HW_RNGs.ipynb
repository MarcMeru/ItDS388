```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import time
%matplotlib inline


def lcg(x0, n):
    a = 1664525
    c = 1013904223
    m = 2 ** 32
    
    output = []
    x = x0
    
    for i in range(n):
        x = (a * x + c) % m
        output.append(x / m)
        
    return output


x0 = time.time()
sample = lcg(x0, 10000)

print('Current time = {0}'.format(x0))
plt.hist(sample, color ='g', edgecolor = 'purple',linewidth = 4)
plt.grid(True)
```
